**h=W*X + b**

```python
x_train=df['ENGINESIZE'].values
y_train=df["CO2EMISSIONS"].values
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
W=tf.Variable(20.0)
b=tf.Variable(30.0)

def h(x):
    y= W*x +b
    return y
h(x_train)
```

```python
def cost_function(y,y_pred):
    out=tf.reduce_mean(tf.square(y-y_pred))
    
    return out
```

# Gradient Decsent

W_new = W_old - alpha * dervitive

```python
import tensorflow as tf

# Initialization of variables
b = tf.Variable(30.0, dtype=tf.float32)  # Setting b to 30.0
w = tf.Variable(tf.random.normal(shape=(), dtype=tf.float32))  # Random weight initialization

alpha = 0.01  # Learning rate
training_epochs = 150

loss_values = []
b_values = []
w_values = []


# Training loop
for epoch in range(training_epochs):
    with tf.GradientTape() as tape:
        y_pred = h(x_train)  # Forward pass: predict y
        loss_value = cost_function(y_train, y_pred)  # Compute loss
        loss_values.append(loss_value.numpy())  # Store loss value

    # Compute gradients of the loss w.r.t. b and w
    gradients = tape.gradient(loss_value, [b, w])

    # Print b value
    print(f'b value = {b.numpy()}')

    # Update variables (gradient descent step)
    b.assign_sub(gradients[0] * alpha)
    w.assign_sub(gradients[1] * alpha)

    # Store the current values of b and w
    b_values.append(b.numpy())
    w_values.append(w.numpy())
```

```python
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.plot(loss_values,'ro')
```

```python
plt.figure(figsize = (10,8))
plt.scatter(x_train,y_train, color = 'green')


for w_plot,b_plot in zip(w_values[0:len(w_values)],b_values[0:len(b_values)]):

  plt.plot(x_train,w_plot*x_train+ b_plot, color = 'red', linestyle = 'dashed')

plt.plot(x_train,w_values[-1]*x_train+ b_values[-1], color = 'black')

```

```python
x_train = df[['ENGINESIZE', 'CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']].values

y_train = df['CO2EMISSIONS'].values
```

```python
model=Sequential()
model.add(Dense(20,activation="relu",input_dim=4))
model.add(Dense(10,activation="relu",))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam" , loss="mse" , metrics=["mse"])
history=model.fit(x_train , y_train ,epochs=100 , verbose=2 , validation_split=0.1)
```
